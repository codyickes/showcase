```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
# View potential columns that need one-hot encoding
one_hot_cols = [col for col in df.columns if df[col].nunique() > 2]
print(one_hot_cols)
```

```python
for col in [
    "Order",
    "PID",
    "MS SubClass",
    "Lot Frontage",
    "Lot Area",
    "Overall Qual",
    "Overall Cond",
    "Year Built",
    "Year Remod/Add",
    "Mas Vnr Area",
    "Exter Qual",  # Need to convert to numeric
    "Exter Cond",  # Need to convert to numeric
    "Bsmt Qual",  # Need to convert to numeric
    "Bsmt Cond",  # Need to convert to numeric
    "Bsmt Exposure",  # Need to convert to numeric
    "BsmtFin SF 1",
    "BsmtFin SF 2",
    "Bsmt Unf SF",
    "Total Bsmt SF",
    "Heating QC",  # Need to convert to numeric
    "1st Flr SF",
    "2nd Flr SF",
    "Low Qual Fin SF",
    "Gr Liv Area",
    "Bsmt Full Bath",
    "Bsmt Half Bath",
    "Full Bath",
    "Half Bath",
    "Bedroom AbvGr",
    "Kitchen AbvGr",
    "Kitchen Qual",  # Need to convert to numeric
    "Functional",  # Need to convert to numeric
    "TotRms AbvGrd",
    "Fireplaces",
    "Fireplace Qu",  # Need to convert to numeric
    "Garage Yr Blt",
    "Garage Cars",
    "Garage Area",
    "Garage Qual",  # Need to convert to numeric
    "Garage Cond",  # Need to convert to numeric
    "Paved Drive",  # Need to convert to numeric or add back for one-hot
    "Wood Deck SF",
    "Open Porch SF",
    "Enclosed Porch",
    "3Ssn Porch",
    "Screen Porch",
    "Pool Area",
    "Pool QC",  # Need to convert to numeric or add back for one-hot
    "Misc Val",
    "Mo Sold",
    "Yr Sold",
    "SalePrice",  # Label (y), shouldn't be included in X, therefore, no One-Hot encoding
    "BsmtFin Type 1",  # Need to convert to numeric
    "BsmtFin Type 2",  # Need to convert to numeric
    "Garage Finish",  # Need to convert to numeric
    "Lot Shape",  # Need to convert to numeric
    "Land Slope",  # Need to convert to numeric
    "Utilities",  # Need to convert to numeric
    "Electrical",  # Need to convert to numeric
    "Fence",  # Need to convert to numeric
]:
    one_hot_cols.remove(col)
```

```python
import numpy as np
```

```python
quality_mapping = {
    "Ex": 4,
    "Gd": 3,
    "TA": 2,
    "Fa": 1,
    "Po": 0,
}

quality_mapping_with_nan = {
    "Ex": 5,
    "Gd": 4,
    "TA": 3,
    "Fa": 2,
    "Po": 1,
    np.nan: 0,  # House feature with quality likely doesn't exist
}

functional_mapping = {
    "Typ": 7,
    "Min1": 6,
    "Min2": 5,
    "Mod": 4,
    "Maj1": 3,
    "Maj2": 2,
    "Sev": 1,
    "Sal": 0,
}

exposure_mapping = {
    "Gd": 4,
    "Av": 3,
    "Mn": 2,
    "No": 1,
    np.nan: 0,
}

bsmtfin_mapping = {
    "GLQ": 6,
    "ALQ": 5,
    "BLQ": 4,
    "Rec": 3,
    "LwQ": 2,
    "Unf": 1,
    np.nan: 0,
}

garage_finish_mapping = {
    "Fin": 3,
    "RFn": 2,
    "Unf": 1,
    np.nan: 0,
}

lot_shape_mapping = {
    "Reg": 3,
    "IR1": 2,
    "IR2": 1,
    "IR3": 0,
}

land_slope_mapping = {
    "Gtl": 2,
    "Mod": 1,
    "Sev": 0,
}

utilities_mapping = {
    "AllPub": 2,
    "NoSewr": 1,
    "NoSeWa": 0,
}

electrical_mapping = {
    "SBrkr": 4,
    "FuseA": 3,
    "FuseF": 2,
    "FuseP": 1,
    "Mix": 0,
    np.nan: 4,
}

fence_mapping = {
    "GdPrv": 4,
    "MnPrv": 3,
    "GdWo": 2,
    "MnWw": 1,
    np.nan: 0,
}
```

```python
qual_cols = [
    "Exter Qual",
    "Exter Cond",
    "Heating QC",
    "Kitchen Qual",
]

qual_nan_cols = [
    "Bsmt Qual",
    "Bsmt Cond",
    "Fireplace Qu",
    "Garage Qual",
    "Garage Cond",
    "Pool QC",
]

# Map strings to ordinal values
df[qual_cols] = df[qual_cols].replace(quality_mapping)
df[qual_cols] = df[qual_cols].apply(pd.to_numeric)
df[qual_nan_cols] = df[qual_nan_cols].replace(quality_mapping_with_nan)
df[qual_nan_cols] = df[qual_nan_cols].apply(pd.to_numeric)
df["Bsmt Exposure"] = df["Bsmt Exposure"].map(exposure_mapping)
df["Functional"] = df["Functional"].map(functional_mapping)
df["BsmtFin Type 1"] = df["BsmtFin Type 1"].map(bsmtfin_mapping)
df["BsmtFin Type 2"] = df["BsmtFin Type 2"].map(bsmtfin_mapping)
df["Garage Finish"] = df["Garage Finish"].map(garage_finish_mapping)
df["Lot Shape"] = df["Lot Shape"].map(lot_shape_mapping)
df["Land Slope"] = df["Land Slope"].map(land_slope_mapping)
df["Utilities"] = df["Utilities"].map(utilities_mapping)
df["Electrical"] = df["Electrical"].map(electrical_mapping)
df["Paved Drive"] = df["Paved Drive"].map({"Y": 2, "P": 1, "N": 0})
df["Fence"] = df["Fence"].map(fence_mapping)

# Convert Street and Central Air to numbers
df["Street"] = df["Street"].map({"Grvl": 0, "Pave": 1})
df["Central Air"] = df["Central Air"].map({"N": 0, "Y": 1})

# Handle NaN for Alley and add back to One-Hot encoding list
df["Alley"] = df["Alley"].fillna("No")
one_hot_cols.append("Alley")
```

```python
# This will save the converted CSV, the One-Hot columns need to be encoded before training
df.to_csv("data/processed_dataset.csv", index=False)
# Columns to One-Hot
one_hot_cols
```
